```python
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

# this time use a scatter plot
plt.scatter(x_values, y_values, s=100)

# give titles and label the axes
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# set the size of tick labels
plt.tick_params(axis='both', labelsize=14)

plt.show()
```

```python
import matplotlib.pyplot as plt

# this time let's use loops to calculate a large range of values
x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values, y_values, s=40)

# give titles and label the axes
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)

# set the size of tick labels
plt.tick_params(axis='both', labelsize=14)

# set the range for each axis
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

# this time remove the edgecolor for more viewing clarity
# also change the plot color to red
plt.scatter(x_values, y_values, c="red", edgecolor="none", s=40)

plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.tick_params(axis='both', labelsize=14)
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

# create a colormap to use a gradient based on a points y-value
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor="none", s=40)

plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Square of Value", fontsize=14)
plt.tick_params(axis='both', labelsize=14)
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```python
"""
Create a plot showing the first 5000 cubic numbers and use a colormap to display to the values.
"""

import matplotlib.pyplot as plt

# x values up to 5000
x_values = list(range(1, 5001))
# y values hold the cubes of the x values
y_values = [x**3 for x in x_values]

# create a colormap to use a gradient based on a points y-value
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.plasma, edgecolor="none", s=40)

plt.title("Cube Numbers", fontsize=24)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Cube of Value", fontsize=14)
plt.tick_params(axis='both', labelsize=14)
plt.axis([0, 5000, 0, 125000000000])

plt.show()
```
